```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
file = 'https://data.cdc.gov/resource/9mfq-cb36.json'
pd.set_option("display.max_rows", None)
df = pd.read_json(file)
df.sort_values(by=['submission_date'], inplace=True, ignore_index=True)
df = df[df['state'] == "AK"]
df
```

```python
file = 'https://data.cdc.gov/resource/9mfq-cb36.json'
df = pd.read_json(file)
df['tot_cases'] = df['tot_cases'].fillna(0)
cases_df = df[['submission_date', 'state', 'tot_cases']].copy()
cases_df.sort_values(by=['state','submission_date'], inplace=True, ignore_index=True)
cases_df['moving'] = cases_df.groupby('state')['tot_cases'].transform(lambda x: x.rolling(7).mean())
# cases_df = cases_df[cases_df['state'] == "CA"]
cases_df['moving'] = cases_df['moving'].fillna(0)
cases_df.sort_values(by=['submission_date'], inplace=True, ignore_index=True)
cases_df
```

```python
# color_discrete_map2 = {'Democrat': 'rgb(0,0,255)', 'Republican': 'rgb(225,0,0)'}
fig = px.choropleth(cases_df,
                    scope = 'usa',
                    locationmode = 'USA-states',
                    locations = 'state',
                    color = 'moving',       
                    hover_name = 'state',
                    hover_data = {'state':False,'tot_cases':True,'moving':True},
                    animation_frame="submission_date",
                    title = 'Popular Vote 1976 - 2016 Presidential Elections'
                   )

fig.show()
```
