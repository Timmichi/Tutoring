# Ex 7.18 Mode of an Array
Since the book was published, NumPy added a median function but still does not have a mode function. As described in the textbook, write a function that will calculate the mode of an NumPy array. Note the following requirements:
* DO NOT use a mode function from another library
* The function should be called `arrmode` (use the started code below)
* It should take one argument, a NumPy array
* It should work for any shape array
* If there is no mode, e.g. all members of the array occur the same amount, show message 'No Mode'.
* Module should return just the mode, not the count.
* If the distribution has several modes, return all of them in a single list.


```python
import numpy as np
```

```python
# 7.18 Function to Calculate Mode of an Array
def arrmode(nparray):
    '''Returns the mode (or modes) of a NumPy array'''
    # Your code here
    counts = np.unique(nparray, return_counts = True)
    print(counts)
    maximum_indices = np.where(counts[1] == max(counts[1]))
    maximum_indices = maximum_indices[0]
    if len(maximum_indices)==1:
        return counts[0][maximum_indices[0]]
    mode= []
    if len(maximum_indices) == len(counts[1]):
        return "No Mode"
    for i in maximum_indices:
        mode.append(counts[0][i])
    return mode # change this line to return the mode
```

```python
# Test 1 - Do not change this code, run it to check your code

array1 = np.array([1,2,3,4,2,2,3])
print("The mode is: ", arrmode(array1))

# Output should be: The mode is: 2
```

```python
# Test 2 - Do not change this code, run it to check your code

array2 = np.array([[1,2,3,4,3],[2,3,3,1,4]])
print("The mode is: ", arrmode(array2))

# Output should be: The mode is: 2
```

```python
# Test 3 - Do not change this code, run it to check your code

array3 = np.array([1,2,3,4,2,2,3,3,5,6])
print("The mode is: ", arrmode(array3))

# Output should be: The mode is: [2,3]
```
