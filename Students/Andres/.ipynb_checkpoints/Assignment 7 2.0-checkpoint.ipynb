```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import ssl
```

```python
ssl._create_default_https_context = ssl._create_unverified_context
pd.set_option("display.max_columns", None)
df = pd.read_csv('https://covidtracking.com/api/v1/states/daily.csv')
df.sort_values(by=['date'], inplace=True, ignore_index=True)
df['positive'] = df['positive'].fillna(0)
cases_df = df[['date', 'state', 'positive']].copy()
cases_df.sort_values(by=['state','date'], inplace=True, ignore_index=True)
cases_df['moving'] = cases_df.groupby('state')['positive'].transform(lambda x: x.rolling(7).mean())
cases_df['moving'] = cases_df['moving'].fillna(0)
cases_df = cases_df[cases_df['state'] == "AK"]
cases_df
```
