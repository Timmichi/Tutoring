1. A box and whisker plot showing multiple box and whisker distributions for each of the four iris traits: sepal length, sepal width, petal length, and petal width.

```python
# Importing the libraries. Do not change this cell! 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
path_to_iris_file = "HW7_8_iris.csv"

df   = pd.read_csv(path_to_iris_file, sep=',',header=None)
print(df)
prev_data = np.array(df.values)
data = np.empty([5, 151], dtype=object)
for i in range(1,6):
    data[i-1] = prev_data[:, i]
```

```python
#TODO: 
#Before you move on, check that your data array has shape 5 x 151
#All five rows should start with the trait name (sepal length, sepal width, petal length, petal width, species)
#All five rows should then contain the values for that trait for a given flower. These might be numbers or strings.

pd.DataFrame(data) #this will help you look at your data values and confirm they're in the right shape. 
```

```python
#TODO create 5 NUMPY arrays, one for each trait
#For each array, remove the first entry that contains the trait name, 
#so that each array is only a set of values for each flower.
#For the arrays with numerical values, cast them to floats. 

sepal_length = np.array(list(map(lambda d: float(d), data[0][1:])))
sepal_width  = np.array(list(map(lambda d: float(d), data[1][1:])))
petal_length = np.array(list(map(lambda d: float(d), data[2][1:])))
petal_width  = np.array(list(map(lambda d: float(d), data[3][1:])))
species      = data[4][1:]
print(species)
```

```python
# TODO:
# Create a box and whisker plot (boxplot) for sepal length, sepal width, petal length, petal width
# Do this all on one boxplot! Do not create multiple plots. 
# Don't forget to label your plot and your axes!
fig = plt.figure(figsize =(10, 7))

# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot([sepal_length, sepal_width, petal_length, petal_width])
 
# show plot
plt.show()
```

```python
# TODO:
# Create a scatter plot with sepal length on the x axis and petal length on the y axis. 
# Color the points using three colors, one for each species of iris. 
# Don't forget to label your axes and your plot!    

# we can't use cmap on discrete parameters. we can only use on numbers/floats.
dictionary = {"setosa": "red", "versicolor": "blue", "virginica": "green"}
for index, flower in enumerate(species):
    species[index] = dictionary[flower]
    
plt.scatter(sepal_length, petal_length, s=100, c=species)


```
