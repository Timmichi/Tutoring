```python
# Ex 5.11 Part 1
# Use a LIST COMPREHENSION to create the LIST of lower case alphabet letters that can be iterated over 
# Name the list ALPHABET
# When you output the list, it will similar to:
# ['a', 'b','c','d', ....... 'z']

# DO NOT type out the list. Instead us a single line List Comprehension to convert 
#  the numbers 97 through 122 to characters
# Output your list to confirm it created the full lower case alphabet
alphabet = [chr(x) for x in range(97,123)]
alphabet
```

```python
# Ex 5.11 Part 2
# Do not use the COUNT function or a dictionary. This part of the assignment tests
# function creation, iterating through a string, use of 'in' operator, 
# creating a new string of unique letters and corresponding counts, and converting to tuples

# Create a function that takes a string as a parameter,
# then summarizes is it by iterating through the string and 
# counting the occurrences of each letter and returns a list of the tuples
# For instance - if you gave it the word 'Apple', it would return the following:
# [('a', 1), ('e', 1), ('l', 1), ('p', 2)]
# Note this is one list made up of several tuple pairs

# Call your function summarize_letters
# It should take one argument, a string

def summarize_letters(string):
    alphabet = [chr(x) for x in range(97,123)]
    counts = [0] * 26
    for letter in string:
        letter = letter.lower()
        indx = alphabet.index(letter)
        counts[indx] += 1
    freq_list = list(zip(alphabet, counts))
    freq_list = list(filter(lambda letter_tuple: letter_tuple[1] > 0, freq_list))
    return freq_list
```

```python
# Test the function
# Should look like: [('a', 1), ('e', 1), ('l', 1), ('p', 2)]
print(summarize_letters("Apple"))
```

```python
# Ex 5.11 Part 3
# Get the string from the user (or hard code it for now)
# Pass it to the summarization function above and receive
# the returned lists.
# Display all output and messages


# teststring  = 'Hello'   # panagram and non-panagram

teststring = "The quick brown fox jumps over the lazy dog."
for letter, freq in summarize_letters("".join(filter(lambda letter: letter in alphabet, teststring.lower()))): print(f"{letter}:{freq}", end=" ")




```

```python
# Step 1:
# Insert 20 random letters in the range 'a' through 'f' into a list
# Use a list comprehension to create the list of 20 random letters
# Print the list left to right (horizontally, not vertically)
import random
randomletters = [chr(random.randrange(ord('a'), ord('f'))) for x in range(1,21)]
print(randomletters,)
```

```python
# Step 2
# Sort the list in ascending order and print out left to right
randomletters.sort()
print(randomletters,)
```

```python
# Sort the list in descending order and print out left to right
randomletters.sort(reverse=True)
print(randomletters,)
```

```python
# Get the unique values, sort them in asending order, 
# and print out left to right
uniqueList=[]
for element in randomletters:
    if element not in uniqueList:
        uniqueList.append(element)
#sort list
print(" ".join(sorted(uniqueList)))

```

## Summing the Triples of the Even Integers from 1 through 10 (using `filter`, `map`, `sum`)

See the textbook for details. Suggested cells are given but feel free to change as you wish to meet the requirements of the assignment. Do not overthink this. There are only two lines of code. The first line is given by me. Note the triple of 2 is 6, e.g., 2 * 3.

Assignment Competency Objectives:
* Create a list through simple iteration
* Use lambdas, map, filter and sum to create that same list
* Use a list comprehension to create the same list
* **The correct answer is 90**  (6 + 12 + 18 + 24 + 30 = 90)

```python
# Step 1: Create the list

numbers = list(range(1, 11))
numbers
```

```python
# Approach A: Use filter, map and sum to calculate the total of the 
# triples of the even integers in the numbers list (one line of code)
even_numbers = list(filter(lambda x: x%2==0, numbers))   #filtering out the even numbers using filter()
triples=list(map(lambda x:x*3 ,even_numbers))   #calculating the triples of each even number using map
total_triples=sum(triples)   #calculatting the sum

print(even_numbers)
print(triples)
total_triples

```

```python
# Approach B: Reimplement the previous line of code with a list comprehension

even_numbers=[x for x in numbers if x%2==0]
triples=[x*3 for x in even_numbers]
total_triples = sum(triples)
print(even_numbers)
print(triples)
print(total_triples)

```
